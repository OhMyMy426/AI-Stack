```python
"""whisper-1 OpenAI"""
from openai import OpenAI
import os
from dotenv import load_dotenv
import time

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY1"))

filename = os.path.join(os.getcwd(), "inputs", "dummy-MP3.mp3")

start = time.time()
with open(filename, "rb") as file:
    response = client.audio.transcriptions.create(
        model="whisper-1",
        file=file,
        response_format="verbose_json",  # or "text"
    )
    print(response.text)
print(f"✅ Total time: {time.time() - start:.2f} seconds")
```

```python
"""GOOGLE TEXT-TO-SPEECH"""

import os
import time
from google.cloud import speech
from dotenv import load_dotenv

load_dotenv()
path = os.getenv("GOOGLE_APPLICATION_CREDENTIALS")
# print("Path exists:", os.path.exists(path))
# print(path)

filename = os.path.join(os.getcwd(), "inputs", "dummy-MP3 (mp3cut.net).mp3")

client = speech.SpeechClient()

with open(filename, "rb") as audio_file:
    content = audio_file.read()

audio = speech.RecognitionAudio(content=content)

config = speech.RecognitionConfig(
    encoding=speech.RecognitionConfig.AudioEncoding.MP3,
    sample_rate_hertz=16000,
    language_code="en-US"
)

operation = client.long_running_recognize(config=config, audio=audio)

print("Processing audio (this may take a while)...")
response = operation.result(timeout=600)

for result in response.results:
    print("Transcript:", result.alternatives[0].transcript)

print(f"Total time for recognition: {time.time() - start:.2f} seconds")
```

```python
"""console.deepgram.com """
import os
import time
from deepgram import DeepgramClient, PrerecordedOptions
from dotenv import load_dotenv

# Start timer
start = time.time()

# Load API key from .env
load_dotenv()
DEEPGRAM_API_KEY = os.getenv("DEEPGRAM_API_KEY")

# Confirm key is loaded
if not DEEPGRAM_API_KEY:
    raise ValueError("Missing DEEPGRAM_API_KEY in environment")

# Prepare file path
FILE = os.path.join(os.getcwd(), "inputs", "dummy-MP3.mp3")

# Initialize Deepgram client
deepgram = DeepgramClient(DEEPGRAM_API_KEY)

# Read audio and send transcription request
with open(FILE, "rb") as buffer_data:
    payload = {"buffer": buffer_data}

    options = PrerecordedOptions(
        smart_format=True,
        model="nova-2",  # You can also try "base" if nova-2 has issues
        language="en-US",  # Change to "zh" or "de" if needed
        # punctuate=True          # Optional, adds punctuation
    )

    response = deepgram.listen.prerecorded.v('1').transcribe_file(payload, options)
    print(response.to_json(indent=4))

# Time summary
print(f"Total time for transcription: {time.time() - start:.2f} seconds")
```

```python
"""google Text-to-Speech AI"""

import os
from google.cloud import texttospeech
from dotenv import load_dotenv

load_dotenv()

input_file = os.path.join(os.getcwd(), "inputs", "content.txt")

client = texttospeech.TextToSpeechClient()

start = time.time()

with open(input_file, "r") as content:
    text = content.read()

    synthesis_input = texttospeech.SynthesisInput(text=text)

    voice = texttospeech.VoiceSelectionParams(
        language_code="en-US",
        name="en-US-Wavenet-D",
        ssml_gender=texttospeech.SsmlVoiceGender.NEUTRAL
    )

    audio_config = texttospeech.AudioConfig(
        audio_encoding=texttospeech.AudioEncoding.MP3
    )

    # call the API
    response = client.synthesize_speech(
        input=synthesis_input,
        voice=voice,
        audio_config=audio_config
    )

with open(os.path.join("outputs", "google_speech_output.mp3"), "wb") as out:
    out.write(response.audio_content)
    print(f"✅ Audio content written to 'output.mp3' within {time.time() - start:.2f} seconds")
```

```python
import openai
import os
import time
from dotenv import load_dotenv

load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY1")
# print(os.getenv("OPENAI_API_KEY1"))
input_file = os.path.join(os.getcwd(), "inputs", "content.txt")

start = time.time()
with open(input_file, "r") as content:
    text = content.read()
    response = openai.audio.speech.create(
        model="tts-1",  # or "tts-1-hd"
        voice="alloy",  # or: alloy, echo, fable, onyx, nova, shimmer
        input=text
    )

with open(os.path.join("outputs", "openai_dummy_output.mp3"), "wb") as f:
    f.write(response.content)

print(f"✅ Audio saved to openai_output.mp3 within {time.time() - start:.2f} seconds")
```

```python
from elevenlabs import VoiceSettings
from elevenlabs.client import ElevenLabs
import os
from dotenv import load_dotenv

load_dotenv()

input_file = os.path.join(os.getcwd(), "inputs", "content.txt")

client = ElevenLabs(api_key=os.getenv("ELEVEN_API_KEY"))

start = time.time()

with open(input_file, "r") as content:
    text = content.read()
    response = client.text_to_speech.convert(
        voice_id="pNInz6obpgDQGcFmaJgB",  # Adam voice
        output_format="mp3_22050_32",
        text=text,
        model_id="eleven_turbo_v2_5",  # use turbo model for low latency
        voice_settings=VoiceSettings(
            stability=0.0,
            similarity_boost=1.0,
            style=0.0,
            use_speaker_boost=True,
            speed=1.0,
        ),
    )

# Save to file
with open(os.path.join("outputs", "elevenlabs_output.mp3"), "wb") as f:
    for chunk in response:
        if chunk:
            f.write(chunk)

print(f"✅ Audio saved to elevenlabs_output.mp3 within {time.time() - start:.2f} seconds")
```
